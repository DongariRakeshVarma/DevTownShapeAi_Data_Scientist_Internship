```python
# Import all necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno

%matplotlib inline
```

```python
# Reading data and storing as a DataFrame
df=pd.read_csv("/content/50_Startup.csv")
```

```python
# Displaying data
df
```

```python
# First 5 elements
df.head()
```

```python
# Last 5 elements
df.tail()
```

```python
# Information about data stored
df.info()
```

```python
# Describe data present - statistical information of numerical values
df.describe()
```

```python
# Round values upto 2 decimal places
df['R&D Spend']=df['R&D Spend'].round(2);
df['Administration']=df['Administration'].round(2);
df['Marketing Spend']=df['Marketing Spend'].round(2);
df['Profit']=df['Profit'].round(2);

df
```

**MANAGING NULL VALUES**

```python
# Displaying number of null values per column
df.isnull().sum()
```

```python
# Heatmap to visualise null values

plt.figure(figsize=(10,10))
sns.heatmap(data=df.isnull(),cmap='viridis_r',linewidths=.25)
plt.xlabel('COLUMN')
plt.ylabel('INDEX')
plt.show()
```

```python
# Plotting number of null values column wise using missingno library

msno.bar(df)
plt.show()
```

```python
# Plotting number of null values as heatmap using missingno library

msno.heatmap(df)
plt.show()
```

```python
# Filling rows having 'R&D Spend' , 'Administration' and 'Marketing Spend' as null values with zeroes

df['R&D Spend'].fillna(0,inplace=True)
df['Administration'].fillna(0,inplace=True)
df['Marketing Spend'].fillna(0,inplace=True)

df
```

```python
# Interpolating 'Profit' values

df['Profit'].interpolate(method='linear',inplace=True)
df['Profit']=df["Profit"].round(2)
df
```

```python
# Deleting rows having 'State' as a null value

df.dropna(inplace=True,axis=0)
df
```

```python
# Checking whether all null values have been removed or not

df.isnull().sum()
```

# VISUALISING DATA

```python
# Plotting 'State' v/s frequency plot 

plt.figure(figsize=(5,5))
base_1=sns.color_palette('muted')
sns.countplot(df['State'],palette=base_1)
plt.ylabel('Number of Startups')
plt.show()
```

```python
# Mean, Max 'R&D Spend' state-wise

RandD=pd.DataFrame({'Mean':df.groupby(['State'])['R&D Spend'].mean().round(2),
                    'Max':df.groupby(['State'])['R&D Spend'].max()})
RandD
```

```python
# Plotting Mean, Max 'R&D Spend' state-wise

plt.figure(figsize=(7,7))
RandD.plot(kind='bar')
plt.show()
```

```python
# Mean, Max 'Administration' state-wise

Administration=pd.DataFrame({'Mean':df.groupby(['State'])['Administration'].mean().round(2),
                    'Max':df.groupby(['State'])['Administration'].max()})
Administration
```

```python
# Plotting Mean, Max and 'Administration' state-wise

plt.figure(figsize=(20,20))
Administration.plot(kind='bar')
plt.show()
```

```python
# Mean, Max and 'Marketing Spend' state-wise

Marketing=pd.DataFrame({'Mean':df.groupby(['State'])['Marketing Spend'].mean().round(2),
                    'Max':df.groupby(['State'])['Marketing Spend'].max()})
Marketing
```

```python
# Plotting Mean, Max 'Marketing Spend' state-wise

plt.figure(figsize=(7,7))
Marketing.plot(kind='bar')
plt.show()
```

```python
# Mean, Max and Min 'Profit' state-wise

Profit=pd.DataFrame({'Mean':df.groupby(['State'])['Profit'].mean().round(2),
                    'Max':df.groupby(['State'])['Profit'].max(),
                    'Min':df.groupby(['State'])['Profit'].min()})
Profit
```

```python
# Plotting Mean, Max and Min 'Profit' state-wise

plt.figure(figsize=(7,7))
Profit.plot(kind='bar')
plt.show()
```

```python
# Plotting 'Profit' earned on different amounts of investment - Line Plot

plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
sns.lineplot(data=df,x="Profit", y="R&D Spend", hue="State")

plt.subplot(1,3,2)
sns.lineplot(data=df,x="Profit", y="Administration", hue="State")

plt.subplot(1,3,3)
sns.lineplot(data=df,x="Profit", y="Marketing Spend", hue="State")
```

```python
# Plotting 'Profit' earned on different amounts of investment - Scatter Plot

plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
sns.scatterplot(data=df,x="Profit", y="R&D Spend", hue="State")

plt.subplot(1,3,2)
sns.scatterplot(data=df,x="Profit", y="Administration", hue="State")

plt.subplot(1,3,3)
sns.scatterplot(data=df,x="Profit", y="Marketing Spend", hue="State")
```

```python
# Calculating total expenditure of each startup

def exp(x):
    return x['R&D Spend'] + x['Administration'] + x['Marketing Spend']

df['Expenditure']=df.apply(exp,axis=1)
df
```

```python
# Plot 'Expenditure' v/s 'Profit' graph

plt.figure(figsize=(10,10))
plt.hist2d(data=df,x='Expenditure',y='Profit',cmap='flare_r')
plt.colorbar()
plt.xlabel('Expenditure')
plt.ylabel('Profit')
plt.show()
```

```python
 # Plot 'Expenditure' v/s 'Profit' graph state-wise

plt.figure(figsize=(20,10))
sns.lineplot(data=df,y="Profit", x="Expenditure", hue="State")
plt.show()
```

```python

# Calculating % of 'Profit' earned 
df['Profit_%']=((df['Profit']/df['Expenditure'])*100).round()
df
```

```python
# Plotting 'Profit_%' as a pie chart

plt.figure(figsize=(8,8))
plt.pie(df['Profit_%'].value_counts(),labels=df['Profit_%'].value_counts().index,startangle=90,counterclock=False)
plt.axis('square')
plt.show()
```
